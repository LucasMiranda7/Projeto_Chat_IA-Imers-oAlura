```python
# Instalando o SDK do Google Generative AI
!pip install -q -U google-generativeai

# Configurações iniciais
import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('SECREAT_KEY')
genai.configure(api_key=api_key)

generation_config = {
  "candidate_count": 1,
  "temperature": 0.5,
}

safety_settings={
    'HATE': 'BLOCK_NONE',
    'HARASSMENT': 'BLOCK_NONE',
    'SEXUAL' : 'BLOCK_NONE',
    'DANGEROUS' : 'BLOCK_NONE'
    }

model = genai.GenerativeModel(model_name='gemini-1.0-pro',
                                  generation_config=generation_config,
                                  safety_settings=safety_settings,)

# Funções para o Projeto Mesa Compartilhada de Pessoas em situações de rua
def doacao_alimentos(tipo_alimento, quantidade, localizacao):
  # Simulação de busca por ponto de coleta (pode ser substituido por lógica real)
  ponto_coleta = "Rua Exemplo, 123 - Centro"
  resposta = f"Obrigado pela sua doação de {quantidade} de {tipo_alimento}! O ponto de coleta mais próximo de {localizacao} é {ponto_coleta}."
  return resposta

def buscar_alimentos(necessidades, localizacao):
  # Simulação de busca por alimentos
  alimentos_disponiveis = "Cestas básicas com arroz, feijão, macarrão e Muito Mais."
  resposta = f"Em {localizacao}, temos disponíveis {alimentos_disponiveis}."
  return resposta

def dicas_nutricao():
  # Simulação de dicas de nutrição (substitua por lógica real)
  dicas = "Consuma arroz, feijão e verduras, beba bastante água e evite alimentos processados que prejudiquem sua saúde."
  return dicas

  #Função para informações para doações a pessoas em situação de Rua
def info_inseguranca_alimentar(doacoes, orgs):
  mensagem = """
  Juntos contra a Fome e a Invisibilidade: Uma Campanha para a Dignidade Humana
Em um mundo de abundância, a fome e a falta de moradia ainda assombram milhões de pessoas.

 No Brasil, a realidade é ainda mais cruel:

 Mais de 30 milhões de pessoas vivem em situação de insegurança alimentar.
 Milhares de pessoas dormem nas ruas, sem ter um lugar para chamar de lar.
 Essa é uma realidade que não podemos aceitar!

 É hora de unirmos forças para construir um futuro mais justo e digno para todos.

 Junte-se a nós nesta campanha e faça a diferença:
         **Doe nas Orgs:
         *Cruz Vermelha:
         Site:[https://www.cruzvermelha.org.br/pb/]
         *Amigos do Bem:
         Site:[https://www.amigosdobem.org/como-ajudar/]
  """
  return mensagem
  # Chamada da função e impressão do texto
texto_campanha = info_inseguranca_alimentar("doe", "orgs")  # Substitua "doacoes" e "orgs" pelos valores reais
print(texto_campanha)

# Função para informações sobre as enchentes no RS
def info_enchentes_rs():
  mensagem = """
  O Rio Grande do Sul foi recentemente atingido por fortes enchentes, causando danos e desalojando milhares de pessoas.
  Diversas organizações estão recebendo doações para ajudar as vítimas.

  **Aqui estão algumas formas de ajudar:**

  **Doações via Pix:**

  * **Defesa Civil do RS:**
    * Chave Pix: [(CNPJ): 92.958.800/0001-38 / instituição: Banrisul]
  * **Central Única das Favelas (CUFA):**
    * Chave Pix: [doacoes@cufa.org.br]
  * **[Outras organizações com chaves Pix]:**
    * [Lista de outras organizações e suas chaves Pix]

  **Vaquinhas Online:**

  * **[Sos Enchentes]:**
    * Link: [enchentes@vakinha.com.br]
  * **[@gradbrasil]:**
    * Link: [PIX: 54465282000121]
  * **[Outras Vaquinhas]:**
    * [Lista de outras vaquinhas e seus links]

  **Outras formas de ajudar:**

  * **Doação de alimentos, roupas e itens de higiene:**
    * Entre em contato com organizações locais para saber como e onde doar.
  * **Voluntariado:**
    * Ofereça seu tempo e habilidades para ajudar nas ações de socorro e reconstrução.

  **Toda ajuda é bem-vinda!**
  """
  return mensagem
  # Chamada da função e impressão do texto
campanha_rs = info_enchentes_rs()  # Remova os argumentos "enchentes" e "ajudar"
print(campanha_rs)


# Início do Chat
prompt = "Olá! Sou um chatbot do Projeto Mesa Compartilhada, focado em combater a fome, a insegurança alimentar e te atualizar sobre a tragedia de enchetes do Rio grande do sul. Como posso te ajudar?"
chat = model.start_chat(history=[])
response = chat.send_message(prompt)
print("Chatbot:", response.text, '\n')

while True:
  prompt = input('Você: ')
  if prompt.lower() == "fim":
    break

  # Identificação da intenção do usuário
  if "doar" in prompt.lower():
    # Extrair informações da doação
    tipo_alimento = "alimentos"
    quantidade = "alguns"
    localizacao = "sua região"
    response = chat.send_message(doacao_alimentos(tipo_alimento, quantidade, localizacao))
  elif "buscar" in prompt.lower():
    # Extrair informações da busca
    necessidades = "alimentos"
    localizacao = "sua região"
    response = chat.send_message(buscar_alimentos(necessidades, localizacao))
  elif "nutrição" in prompt.lower():
    response = chat.send_message(dicas_nutricao())
  #verificar se o usuario menciounou a inseguranca alimentar


  # Verificar se o usuário mencionou as enchentes
  elif any(palavra in prompt.lower() for palavra in ["enchente", "rs", "rio grande do sul", "ajuda" "orgs" "doacao"]):
    response = chat.send_message(info_enchentes_rs())
    response = chat.send_message(info_inseguranca_alimentar())
  else:
    response = chat.send_message(prompt)


  print("Chatbot:", response.text, '\n')
```
